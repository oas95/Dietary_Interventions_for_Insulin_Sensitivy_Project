# Data Cleaning for Analysis 

```python
#importing dependencies 
import pandas as pd
```

```python
#Gardner Data Cleaning
Ga_df = pd.read_csv('../PyDIonIS/Data/Gardner18_Data/DietaryIntake_Table.csv')
Ga2_df = pd.read_csv('../PyDIonIS/Data/Gardner18_Data/DietaryIntake2_Table.csv')
#Renaming Columns
Ga_df.rename(columns= {'Unnamed: 0': 'Low_Fat',
                       'Low-Fat vs Low-Carbohydrate Diet on Weight Loss in Overweight Adults Original Investigation Research' : 'Low_Carb'},
             inplace=True)
Ga_df.head()
```

```python
#Splitting Information to new Columns
Ga_df = Ga_df.Low_Carb.str.split(" ", expand=True)
#Dropping Rows 
Ga_df = Ga_df.drop([0,1,2,3])
#Dropping Coulumns (8, 9, 10, 11)
Ga_df = Ga_df.drop(columns=[8, 9, 10, 11])
```

```python
Ga_df
```

```python
#shifting specific rows to align with original table (Pretty sure theres a better way to do this )
Ga_df.loc[5] = Ga_df.loc[5].shift(periods=1,axis=0)
Ga_df.loc[10] = Ga_df.loc[10].shift(periods=1,axis=0)
Ga_df.loc[15] = Ga_df.loc[15].shift(periods=1,axis=0)
Ga_df.loc[20] = Ga_df.loc[20].shift(periods=1,axis=0)
Ga_df.loc[25] = Ga_df.loc[25].shift(periods=1,axis=0)
Ga_df.loc[30] = Ga_df.loc[30].shift(periods=1,axis=0)
Ga_df.loc[35] = Ga_df.loc[35].shift(periods=1,axis=0)
Ga_df.loc[40] = Ga_df.loc[40].shift(periods=1,axis=0)
Ga_df.loc[45] = Ga_df.loc[45].shift(periods=1,axis=0)
Ga_df.loc[50] = Ga_df.loc[50].shift(periods=1,axis=0)

```

```python
#Checking if they have shifted
Ga_df.head()
```

```python

#Mergging Columns
Ga_df['Nutrients_Timepoint'] = Ga_df[0] + ' ' + Ga_df[1]
#Renaming 
Ga_df.rename(columns= { 2 : 'Low_Fat_SS',
                       3 : 'Low_Fat_Mean',
                       4 : 'Low_Fat_SD',
                       5 : 'Low_Carb_SS',
                       6 : 'Low_Carb_Mean',
                       7 : 'Low_Carb_SD'},
             inplace=True)
Ga_df.head()
```

```python
#Dropping Non essential columns 
Ga_df = Ga_df.drop(columns= [0 ,1])

Ga_df.head()
```

```python
#Reording Columns
Ga_df = Ga_df[['Nutrients_Timepoint', 'Low_Fat_SS', 'Low_Fat_Mean', 'Low_Fat_SD', 'Low_Carb_SS', 'Low_Carb_Mean', 'Low_Carb_SD']]
#Reseting index
Ga_df = Ga_df.reset_index().drop('index', axis=1)

Ga_df.head()
```

```python
#Updating values/strings in certain cells
Ga_df.loc[0,['Nutrients_Timepoint']] = ['Total Energy Intake, kcal']
Ga_df.loc[1,['Nutrients_Timepoint']] = ['Baseline']
Ga_df.loc[6,['Nutrients_Timepoint']] = ['Baseline']
Ga_df.loc[11,['Nutrients_Timepoint']] = ['Baseline']
Ga_df.loc[16,['Nutrients_Timepoint']] = ['Baseline']
Ga_df.loc[22,['Nutrients_Timepoint']] = ['Baseline']
Ga_df.loc[28,['Nutrients_Timepoint']] = ['Baseline']
Ga_df.loc[34,['Nutrients_Timepoint']] = ['Baseline']
Ga_df.loc[40,['Nutrients_Timepoint']] = ['Baseline']
Ga_df.loc[46,['Nutrients_Timepoint']] = ['Baseline']
Ga_df.loc[0,['Low_Fat_SS']] = ['None']
Ga_df.loc[0,['Low_Fat_Mean']] = ['None']

Ga_df.head()
```

```python
#Stripping parenthesis from values
try:
    Ga_df['Low_Fat_SD'] = Ga_df['Low_Fat_SD'].str.replace(r'\D', '')
except:
    (TypeError)
try:
    Ga_df['Low_Carb_SD'] = Ga_df['Low_Carb_SD'].str.replace(r'\D', '')
except:
    (TypeError)
```
